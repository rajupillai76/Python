this take each of the Geco FastBooking files and converts them to JSON

```python
import pandas as pd
from pandas import DataFrame
import pyarrow as pa
import pyarrow.parquet as pq
from datetime import datetime
import os
import shutil
```

```python
sourcefilefolder = '/Users/raju.pillai/gitrepo/python/data/DataAcquisition/source/geco/'
targetfilefolder = '/Users/raju.pillai/gitrepo/python/data/DataAcquisition/output/json/geco/'
processedfilefolder = '/Users/raju.pillai/gitrepo/python/data/DataAcquisition/source/processed/geco/'
```

```python
for csvfile in os.listdir(sourcefilefolder):
   if csvfile.endswith('.csv'):
       folder_file = (os.path.join(sourcefilefolder, csvfile))
       jsonfile = (os.path.splitext(csvfile)[0])
       fbcsvfile = open(folder_file,'r')
       fbjsonfile = open ((os.path.join(targetfilefolder, jsonfile)) + '.json','w+')
  #     print(os.path.join(targetfilefolder, jsonfile))
       fbjsondf = DataFrame(pd.read_csv(fbcsvfile, sep=';', engine='python'))
       export = fbjsondf.to_json(fbjsonfile,orient='records')
       del fbjsondf
       processed_folder_file = (os.path.join(processedfilefolder, csvfile))
       shutil.move(folder_file,processed_folder_file)
   else:
       print ('Could not loop thru the folder')
```
