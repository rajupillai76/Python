```python
import pandas as pd
from pandas import DataFrame
import pyarrow as pa
import pyarrow.parquet as pq
from datetime import datetime
import sys
import csv
csv.field_size_limit(sys.maxsize)
```

```python
chainfile = open('/Users/raju.pillai/gitrepo/python/data/DataAcquisition/source/dimensions/chain.txt','r')
brandfile = open('/Users/raju.pillai/gitrepo/python/data/DataAcquisition/source/dimensions/brand.txt','r')
propertyfile = open('/Users/raju.pillai/gitrepo/python/data/DataAcquisition/source/dimensions/property.txt','r')
```

```python
chaindata = DataFrame(pd.read_csv(chainfile, sep='|', engine='python'))
branddata = DataFrame(pd.read_csv(brandfile, sep='|', engine='python'))
propertydata = DataFrame(pd.read_csv(propertyfile, sep='|', engine='python'))
```

```python
#print (list(propertydata.columns))
```

```python
chain_brand_combined_df = pd.merge(chaindata, branddata, left_on='chain_code', right_on='chain_code', how='inner')
```

```python
property_chain_brand_combined_df = pd.merge(chain_brand_combined_df, propertydata, left_on='brand', right_on='property_brand', how='inner' )
```

```python
#for col in property_chain_brand_combined_df.columns: 
#   print(col) 
```
