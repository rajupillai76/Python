# this script reads a URL and get data

```python
import urllib.request
```

```python
#goes the the url and get the result code 
webURL = urllib.request.urlopen("http://www.pillai.me")
print ("result code: " + str(webURL.getcode()))
```

```python
#prints the html content of the page
data = webURL.read()
print(data)
```
